#### Forecasting map

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
```

```python
shapefile_path = "J:/USA_yield prediction/forecast/USA_states.shp"
gdf = gpd.read_file(shapefile_path)
gdf.plot(cmap = 'hsv', edgecolor = 'black', column = 'COUNTYNS1')
```

```python
prediction_data = pd.read_csv("J:/USA_yield prediction/forecast/2017/2017y.csv")
merged_data1 = gdf.merge(prediction_data, on="COUNTYNS1", how="left")
# Check the merged data
print(merged_data1.head())
fig1, ax = plt.subplots(figsize=(10, 8))

# Plot counties with yield data
merged_data1.plot(column="Yield", cmap="viridis", linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

# Plot counties without yield data
merged_data1[merged_data1["Yield"].isnull()].plot(facecolor='white', linewidth=0.8, ax=ax, edgecolor='0.8')

# Set map title and legend
ax.set_title("Yield Prediction Map")

# Remove axes
ax.set_axis_off()

# Display the map
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

fig, axes = plt.subplots(1, 5, figsize=(20, 6))
map1 = merged_data1.plot(column="Yield", cmap="viridis", linewidth=0.8, ax=axes[0], edgecolor='0.8', legend=False)
merged_data1[merged_data1["Yield"].isnull()].plot(facecolor='white', linewidth=0.8, ax=axes[0], edgecolor='0.8')
axes[0].grid(color='lightgray', linestyle='--', linewidth=0.2)

map2 = merged_data2.plot(column="Yield", cmap="viridis", linewidth=0.8, ax=axes[1], edgecolor='0.8', legend=False)
merged_data2[merged_data2["Yield"].isnull()].plot(facecolor='white', linewidth=0.8, ax=axes[1], edgecolor='0.8')
axes[1].grid(color='lightgray', linestyle='--', linewidth=0.2)

map3 = merged_data3.plot(column="Yield", cmap="viridis", linewidth=0.8, ax=axes[2], edgecolor='0.8', legend=False)
merged_data3[merged_data3["Yield"].isnull()].plot(facecolor='white', linewidth=0.8, ax=axes[2], edgecolor='0.8')
axes[2].grid(color='lightgray', linestyle='--', linewidth=0.2)

map4 = merged_data4.plot(column="Yield", cmap="viridis", linewidth=0.8, ax=axes[3], edgecolor='0.8', legend=False)
merged_data4[merged_data4["Yield"].isnull()].plot(facecolor='white', linewidth=0.8, ax=axes[3], edgecolor='0.8')
axes[3].grid(color='lightgray', linestyle='--', linewidth=0.2)

map5 = merged_data5.plot(column="Yield", cmap="viridis", linewidth=0.8, ax=axes[4], edgecolor='0.8', legend=False)
merged_data5[merged_data5["Yield"].isnull()].plot(facecolor='white', linewidth=0.8, ax=axes[4], edgecolor='0.8')
axes[4].grid(color='lightgray', linestyle='--', linewidth=0.2)
fig.text(0.5, 0.84, "Forecasting Map", fontsize=16, weight='bold', ha='center')

# Add the horizontal legend to the map4 plot
bounds = [0, 1, 2, 3, 4]  # Define the boundaries for the color categories
colors = plt.cm.viridis(np.linspace(0, 1, len(bounds) - 1))  # Get the colors from the viridis colormap
cmap = mcolors.ListedColormap(colors)  # Create a colormap with the defined colors
norm = mcolors.BoundaryNorm(bounds, cmap.N)  # Create a boundary norm based on the boundaries

cbar_ax = fig.add_axes([0.4, 0, 0.2, 0.03])  # Define the position and size of the legend
cbar = fig.colorbar(plt.cm.ScalarMappable(cmap=cmap, norm=norm), cax=cbar_ax, orientation='horizontal', ticks=bounds, boundaries=bounds)
cbar.set_ticklabels(['0-1 t/ha', '1-2 t/ha', '2-3 t/ha', '3-4 t/ha', '4 t/ha'])
cbar.ax.tick_params(labelsize=13)  # Adjust the tick label size

# Move the legend title to the top
cbar_ax.set_title('Yield Categories', loc='center', pad=10,  fontsize=15)

# Adjust the position of the legend
# Adjust the position of the legend
cbar.ax.xaxis.set_label_coords(0.5, -2)  # Adjust the label position
cbar_ax.xaxis.set_label_position('bottom')
cbar_ax.xaxis.set_ticks_position('bottom')

merged_data4.plot(ax=axes[3], facecolor='none', edgecolor='none')  # Plot an empty polygon for the legend

plt.subplots_adjust(bottom=0.2)  # Adjust the bottom margin to accommodate the legend
# Add years of the map at the bottom
fig.text(0.18, 0.13, '2017', ha='center', fontsize=16)
fig.text(0.35, 0.13, '2018', ha='center', fontsize=16)
fig.text(0.5, 0.13, '2019', ha='center', fontsize=16)
fig.text(0.67, 0.13, '2020', ha='center', fontsize=16)
fig.text(0.84, 0.13, '2021', ha='center', fontsize=16)

# Adjust spacing between subplots
plt.subplots_adjust(top=0.85, bottom=0.15)

# Show the plot
plt.show()
```
